### Now we will check for stationary of data

```python
# Testing for stationary
from statsmodels.tsa.stattools import adfuller
```

```python
oil_result=adfuller(df['Oil Investments'])
```

```python
print("The values given as output by adfuller is : \n 'ADF test static','p-value','#lags used','number of observation used'\n\n",oil_result)
```

```python
gold_result=adfuller(df['Gold Investments'])
```

```python
print("The values given as output by adfuller is : \n 'ADF test static','p-value','#lags used','number of observation used'\n\n",gold_result)
```

```python
comp_result=adfuller(df['Comp Stock'])
```

```python
print("The values given as output by adfuller is : \n 'ADF test static','p-value','#lags used','number of observation used'\n\n",comp_result)
```

```python
othershare_result=adfuller(df['Other sharesInvestments'])
```

```python
print("The values given as output by adfuller is : \n 'ADF test static','p-value','#lags used','number of observation used'\n\n",othershare_result)
```

```python
#Ho: It is non stationary
#H1: It is stationary

def adfuller_test(df):
    result=adfuller(df) # same as above ,variable name changed
    labels = ['ADF Test Statistic','p-value','#Lags Used','Number of Observations Used']
    for value,label in zip(result,labels):
        print(label+' : '+str(value) )
    if result[1] <= 0.05:
        print("strong evidence against the null hypothesis(Ho), reject the null hypothesis. Data  is stationary")
    else:
        print("weak evidence against null hypothesis, time series is non-stationary ")
```

```python
adfuller_test(df['Oil Investments'])
```

```python
adfuller_test(df['Gold Investments'])
```

```python
adfuller_test(df['Comp Stock'])
```

```python
adfuller_test(df['Other sharesInvestments'])
```

## So, our data is stationary
